```python
import pandas as pd

operators = pd.read_csv('/content/1__Operator_Work_Schedule.csv')
skills = pd.read_csv('/content/2__Topic_Interactions_with_Higher_Missing_Rate.csv')
topics = pd.read_csv('/content/3__Hourly_Intensity_by_Category.csv')

operators.set_index('ID', inplace=True)
skills.set_index('ID', inplace=True)
topics.set_index('Category', inplace=True)
```

```python
import numpy as np
import cvxpy as cp

final = np.zeros((operators.shape[0], 10))

for hour in range(24):
  x = cp.Variable((skills.shape[0], 10), integer=True)
  obj = cp.Minimize(cp.sum(cp.multiply(x, np_skills)) - cp.sum(x) * 10000)

  constraints = [
    (cp.sum(cp.multiply(x, np_skills), axis=1) <= 50),
    (cp.sum(x, axis=0) == topics.iloc[:, hour].sum(axis=0)),
    (x >= 0)]

  for i in range(skills.shape[0]):
    if np_operators[i, hour] == 0:
      constraints.append(x[i] == 0)

  for i in range(skills.shape[0]):
    for j in range(2):
      if np_skills[i, j] == -1:
        constraints.append(x[i, j] == 0)

  pp = cp.Problem(obj, constraints)
  pp.solve()

  if str(x.value) != 'None' and str(x.value) != 'inf':
    final = np.add(np.round(x.value, 0), final)

  print(x.value)

  if str(x.value) != 'None' and str(x.value) != 'inf' and hour <= 22:
    topics.iloc[:, hour + 1] += topics.iloc[:, hour].sum(axis=0) - np.sum(x.value, axis=0)

pd.DataFrame(final)
```
